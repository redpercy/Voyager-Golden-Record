```python
#Image Generation On Left and Right Channels

def image_generation_left(Reference_Image_Name, Reference_Initial_Value, Reference_Adjustment_Value):
    Reference_Image_Name = []
    starting_value = Reference_Initial_Value
    pixel_width = 734
    adjustment = Reference_Adjustment_Value

    for read in range(512):
        Reference_Image_Name.append(left_channel[int(starting_value):int(starting_value+pixel_width)])
        starting_value += pixel_width + adjustment

    plt.figure(figsize = (8,12))
    plt.imshow(Reference_Image_Name, cmap = 'gist_yarg', aspect = '2', origin = 'lower')
    plt.show()

def image_generation_right(Reference_Image_Name, Reference_Initial_Value, Reference_Adjustment_Value):
    Reference_Image_Name = []
    starting_value = Reference_Initial_Value
    pixel_width = 734
    adjustment = Reference_Adjustment_Value

    for read in range(512):
        Reference_Image_Name.append(right_channel[int(starting_value):int(starting_value+pixel_width)])
        starting_value += pixel_width + adjustment

    plt.figure(figsize = (8,12))
    plt.imshow(Reference_Image_Name, cmap = 'gist_yarg', aspect = '2', origin = 'lower')
    plt.show()

#Final Function
    
def generate_images():
    
    length_left = len(reference_data_left)
    length_right = len(reference_data_right)
    
    Image_Name_left = reference_data_left['Image_Name']
    Image_Name_right = reference_data_right['Image_Name']
    
    Initial_Values_left = reference_data_left['Initial_Values']
    Initial_Values_right = reference_data_right['Initial_Values']
    
    Adjustment_Values_left = reference_data_left['Adjustment_Values']
    Adjustment_Values_right = reference_data_right['Adjustment_Values']
    
    for index in range(length_left):
        reference_image_name_left = Image_Name_left[index]
        reference_initial_value_left = Initial_Values_left[index]
        reference_adjustment_value_left = Adjustment_Values_left[index]
        
        image_generation_left(reference_image_name_left, reference_initial_value_left, reference_adjustment_value_left)
   
    for index in range(length_right):
        reference_image_name_right = Image_Name_right[index]
        reference_initial_value_right = Initial_Values_right[index]
        reference_adjustment_value_right = Adjustment_Values_right[index]
        
        image_generation_right(reference_image_name_right, reference_initial_value_right, reference_adjustment_value_right)   
    
```
